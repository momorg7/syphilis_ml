Import the necessary packages and mount the drive

```python
import pandas as pd
import numpy as np

from collections import Counter
from imblearn.under_sampling import RandomUnderSampler

from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

#Action

```python
path = 'PATH_TO_DATASET_IN_DRIVE'
dataset_name = 'DATASET_NAME'

df = pd.read_csv(path+dataset_name+".csv", sep=',') # read the dataset csv
```

```python
df.head() # see the head of data
```

```python
X = df.drop(labels='mc_cri_vdrl', axis=1) # separate the target from the attributes
y = df['mc_cri_vdrl']
```

```python
rus = RandomUnderSampler( # apply Random Undersampling
    sampling_strategy='majority', # decrease the majority class
    random_state=42,
    replacement=False
)

X_res, y_res = rus.fit_resample(X, y)
Counter(y_res)
```

```python
X_res.insert(0, 'mc_cri_vdrl', y_res, True)
X_res.head()
```

```python
name_csv = '' # put csv name
X_res.to_csv(path+name_csv, sep=',', index=False)
```
